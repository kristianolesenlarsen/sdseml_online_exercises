```python
import numpy as np
import matplotlib.pyplot as plt
from functools import lru_cache


@lru_cache(maxsize=1000)
def chebyshevs_poly(n,x):
    if n==1:
        return 1
    elif n==2:
        return x
    else:
        return 2*x*chebyshevs_poly(n=n-1,x=x)-chebyshevs_poly(n-2,x=x)
    
```

```python
plt.plot([chebyshevs_poly(n=1,x=x_tmp) for x_tmp in np.linspace(start=-1,stop=1,num=10000)])
plt.plot([chebyshevs_poly(n=2,x=x_tmp) for x_tmp in np.linspace(start=-1,stop=1,num=10000)])
plt.plot([chebyshevs_poly(n=3,x=x_tmp) for x_tmp in np.linspace(start=-1,stop=1,num=10000)])
plt.plot([chebyshevs_poly(n=4,x=x_tmp) for x_tmp in np.linspace(start=-1,stop=1,num=10000)])
plt.plot([chebyshevs_poly(n=5,x=x_tmp) for x_tmp in np.linspace(start=-1,stop=1,num=10000)])
```
